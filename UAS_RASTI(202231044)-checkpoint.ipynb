```python
#Import library
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Membaca gambar
image_path = ('rasti4.jpg')
image = cv2.imread(image_path)
```

```python
# Mengkonversi gambar ke grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Deteksi tepi menggunakan Canny Edge Detection
edges = cv2.Canny(gray_image, 40, 80)
```

```python
# Deteksi kontur
contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Mengkonversi grayscale ke BGR untuk menggambar warna
contour_image = cv2.cvtColor(gray_image, cv2.COLOR_GRAY2BGR)  
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)
```

```python
#Menampikan output 1
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')

axs[1].imshow(edges, cmap='gray')
axs[1].set_title('Canny Edge Detection')

plt.show()
```

```python
#Menampilkan output 2
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')

axs[1].imshow(edges, cmap='gray')
axs[1].set_title('Canny Edge Detection')

axs[2].imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
axs[2].set_title('Contours Detection')

plt.show()
```
